Problem 2:
How many Unique patterns that exist in the historical stock data set, based on
fluctuations in price.

```python
#Percentage change between the current and a prior element
#We are finding ou value for yearly so multiplying with 252
stock_ret = data_new.pct_change().mean()*252 
stock_ret
```

```python
#converting into series to dataframe
stock_return = pd.DataFrame(stock_ret)
stock_return.columns  = ['Stock_Return']
stock_return
```

```python
from math import sqrt
stock_ret_volatility = data_new.pct_change().std()*sqrt(252)
stock_ret_volatility
```

```python
type(stock_ret_volatility)
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
#format the data into numpy array to feed into K-means algorithm. change1D to 2D array
x = np.asarray([np.asarray(stock_ret), np.asarray(stock_ret_volatility)]).T


wcss =[]
for i in range(1,11):
    kmeans = KMeans(n_clusters = i)
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)

#fig = plt.figure(figsize=(15,5))
plt.plot(range(1,11),wcss)
plt.grid(True)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

From the above Elbow method we see that the curve has a steep at cluster no3 or 4

```python
kmeans.cluster_centers_[:,0]
kmeans.cluster_centers_
```

```python
from pylab import plot, show
from scipy.cluster.vq import kmeans,vq
# Computing kmeans with K =4 (4 clusters)
centroids,_= kmeans(x, 5)
# Assign each sample to a cluster
idx,_ = vq(x, centroids)
```

```python
details = [(name,cluster) for name, cluster in zip(stock_return.index,idx)]
details[:5]
```

```python

df = pd.DataFrame(details, columns=['Stock_Name','Cluster_No'])
df.Cluster_No.value_counts()
```

```python
df=pd.DataFrame(data=scaled_data, columns= data_new.columns)
df.head()
```

```python
from sklearn.decomposition import PCA
pca = PCA()
principalComponents = pca.fit_transform(df)
plt.figure()
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('Number of Components')
plt.ylabel('Variance (%)') #for each component
plt.title('Explained Variance')
plt.grid(True)
plt.show()
```

From the diagram above, it can be seen that 25 principal components explain almost 95% of the variance in data.

So, instead of giving all the columns as input, weâ€™d only feed these 25 principal components of the data to the machine learning algorithm and weâ€™d obtain a similar result.

```python
pca = PCA(n_components=25)
new_data = pca.fit_transform(df)
# This will be the new data fed to the algorithm.
principal_Df = pd.DataFrame(pca.components_, columns = data_new.columns)
principal_Df.head()
```
